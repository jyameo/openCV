# Drawing a line

```python
canvas = np.zeros((300, 300, 3), dtype = "uint8")
green = (0, 255, 0)
cv2.line(canvas, (0,0), (300, 300), green)

red = (255, 0, 0)
cv2.line(canvas, (300,0), (0, 300), red)

plt.imshow(canvas)
plt.show()
```

# Drawing a rectangle

```python
green = (0, 255, 0)
red = (255, 0, 0)
blue = (0, 255, 255)

canvas2 = np.zeros((300, 300, 3), dtype = "uint8")
cv2.rectangle(canvas2, (50,200), (200,225), red, 5)
cv2.rectangle(canvas2, (200,50), (225,125), green)
cv2.rectangle(canvas2, (80,125), (150,80), blue, -1)
plt.imshow(canvas2)
plt.show()
```

# Drawing a circle

```python
green = (0, 255, 0)
red = (255, 0, 0)
blue = (0, 255, 255)

canvas3 = np.zeros((300, 300, 3), dtype = "uint8")
(centerX, centerY) = (canvas3.shape[1] // 2, canvas3.shape[0] // 2)
white = (255, 255, 255)

for r in range(0, 175, 5):
    cv2.circle(canvas3, (centerX, centerY), r, white)
    
plt.imshow(canvas3)
plt.show()
```

```python
canvas4 = np.zeros((300, 300, 3), dtype = "uint8")
for i in range(0, 25):
    radius = np.random.randint(5, high = 200)
    color = np.random.randint(0, high = 256, size = (3,)).tolist()
    
    pt = np.random.randint(0, high = 300, size = (2,))
    
    cv2.circle(canvas4, tuple(pt), radius, color, -1)
    
plt.imshow(canvas4)
plt.show()
```
