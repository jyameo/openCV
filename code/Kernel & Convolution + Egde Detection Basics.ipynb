```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np

image =  cv2.imread("/root/openCV/bin/im_test")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

row, col, cha = image.shape
print("Shape: {}".format(image.shape))
print("Width: {} pixels".format(col))
print("Height: {} pixels".format(row))
print("Channels: {} pixels".format(cha))

plt.imshow(image)
plt.title("Original image")
plt.show()
```

# Convolution

```python
kernel_i = np.array(([0,0,0],
                  [0,1,0],
                  [0,0,0]),
                np.float32)
```

```python
kernel_1 = np.ones((3,3), np.float32)
```

```python
kernel_0 = np.zeros((3,3), np.float32)
```

```python
print(kernel_i)
print(kernel_0)
print(kernel_1)
```

## cv2.filter2D(img, -1)

- -1 to keep same depth

```python
output_i = cv2.filter2D(image, -1,kernel_i)
plt.imshow(output_i)
plt.title("OUTPUT_IDENTITY")
plt.show()
```

# Edge detection filter

```python
kernel_ed = np.array((
        [1,0,-1],
        [0,0,0],
        [-1,0,1]),
            np.float32)
```

```python
output_ed = cv2.filter2D(image, -1,kernel_ed)
plt.imshow(output_ed)
plt.title("OUTPUT_EDGE_DETECTION_FILTER")
plt.show()
```

```python
kernel_ed_2 = np.array((
        [-1,-1,-1],
        [-1,8,-1],
        [-1,-1,-1]),
            np.float32)
```

```python
output_ed_2 = cv2.filter2D(image, -1,kernel_ed_2)
plt.imshow(output_ed_2)
plt.title("OUTPUT_EDGE_DETECTION_FILTER_2")
plt.show()
```
