```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
image =  cv2.imread("/root/openCV/bin/im_test")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

row, col, cha = image.shape
print("Shape: {}".format(image.shape))
print("Width: {} pixels".format(col))
print("Height: {} pixels".format(row))
print("Channels: {} pixels".format(cha))

plt.imshow(image)
plt.title("Original image")
plt.show()
```

# Noise

```python
output = np.zeros(image.shape, np.uint8)
```

```python
prob = 0.05
for i in range(row):
    for j in range(col):
        r = np.random.rand()
        if r < prob /2:
            output[i][j] = [0,0,0]
        elif r < prob:
            output[i][j] = [255,255,255]
        else:
            output[i][j] = image[i][j]
```

```python
plt.imshow(output)
plt.title("SALT & PEPPER")
plt.show()
```

SNR = Signal to Noise ratio
SNR = Signal Power/Noise Power
