```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
image =  cv2.imread("/root/openCV/bin/im_test")

print("Shape: {}".format(image.shape))
print("Width: {} pixels".format(image.shape[1]))
print("Height: {} pixels".format(image.shape[0]))
print("Channels: {} pixels".format(image.shape[2]))

plt.imshow(image)
plt.show()
```

# Average Blurring

```python
blurred = np.hstack([
    cv2.blur(image, (3,3)),
    cv2.blur(image, (7,7)),
    cv2.blur(image, (9,9))])
```

```python
plt.imshow(blurred)
plt.show()
```

# Gaussian Blurring

```python
g_blurred = np.hstack([
    cv2.GaussianBlur(image, (3,3), 0),
    cv2.GaussianBlur(image, (7,7), 0),
    cv2.GaussianBlur(image, (9,9), 0)])
```

```python
plt.imshow(g_blurred)
plt.show()
```

# Median Blurring

```python
m_blurred = np.hstack([
    cv2.medianBlur(image, 3),
    cv2.medianBlur(image, 7),
    cv2.medianBlur(image, 9)])
```

```python
plt.imshow(m_blurred)
plt.show()
```

# Bilateral Filter

```python
bf_blurred = np.hstack([
    cv2.bilateralFilter(image, 3, 21, 21),
    cv2.bilateralFilter(image, 7, 31, 31),
    cv2.bilateralFilter(image, 9, 51, 51)])
```

```python
plt.imshow(bf_blurred)
plt.show()
```
